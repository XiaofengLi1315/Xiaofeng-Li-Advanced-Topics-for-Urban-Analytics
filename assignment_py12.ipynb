## Schelling model desciption <a id='1'></a>

The Schelling model is an abstract model that explores the effect of individual relocation behaviours on macroscale segregation.
The Schelling Agent Based Moodel (ABM) includes: a space, and agents and objects that temporally interact with each other and with the model space.
We describe each component of the Schelling ABM hereafter.

### Space
The space is represented as a grid of square cells. Each grid cell represents a spatial unit that one agent can occupy.
Each spatial unit has 8 immediate surrounding neighbours defined as the "Moore neighbourhood".

### Agents
The model includes one type of agent representing an individual or a household.
Each agent has the following state parameters (i.e., variables that dictate the state of the agent during the simulation):
1. `happy?`           ; indicates whether the agent is happy or not (happy agents have a percentage of surrounding similar agents above the `similarity_threshold`)
2. `similar-nearby`   ; the number of neighboring spatial units that have an agent similar to the agent under analysis
3. `other-nearby`     ; the number of neighbouring agents that are different from the agent under analysis
4. `total-nearby`     ; the total numebr of neighbouring agents

Agents aim to settle in a spatial unit that makes them happy by being surrounding with a high ratio of simiar agents in their Moore neighbourhood. If the agent is unhappy (due to being surrounded by a high ratio of agents different from itself), the agent moves from the current spatial unit to another one randomly. The decision tree of the agents can be summarised as follows:
- if `similar-nearby / total-nearby > similarity_threshold` (i.e., surrrounded by many similar agents) --> happy --> remain in current spatial unit
- if `other-nearby / total-nearby > similarity_threshold` (i.e., surrrounded by many different agents) --> unhappy --> move to another random spatial unit

### Temporality
Each time step in the model represents an iteration where each agent checks whether they are happy or not. Following the previously stated decision tree, each agent decides whether it will move or not.

Part 1: What are ABMs, and what are their key features? Cite at least 2 examples.

Agent-Based Models (ABM) are a class of computational simulation models that represent a system as a collection of many autonomous, interacting entities. By simulating the behaviors and interactions of individuals, ABM reproduces the overall dynamics of the system. Unlike traditional top-down modeling approaches, ABM does not rely on aggregate equations to describe the system; instead, it constructs behavioral logic at the individual level. Each agent acts according to its own rules, and system-level patterns emerge from the collective behavior of all agents(Macal and North, 2005). For this reason, ABM is particularly powerful in studying complex adaptive systems, where the whole is often greater than the sum of its parts. Two classic examples of ABM are the Schelling model and the Boids model.

The core characteristics of ABM can be understood from six dimensions. First, each agent is an autonomous decision-making entity with its own internal state and behavioral rules, and agents differ from one another(Bonabeau, 2002). Second, agents’ behaviors are influenced only by their local neighborhoods and they have no knowledge of the overall system. In the Schelling model, each agent evaluates only its eight adjacent cells, and macro-level patterns naturally arise from the accumulation of countless such local responses(Schelling, 1971).

Emergence is another defining feature of ABM: complex macro-level patterns arise from simple micro-level rules. Closely related to emergence is ABM’s explicit treatment of space and time. The model evolves in discrete time steps, and spatial proximity determines who interacts with whom. In the Boids model, each bird adjusts its flight direction based on the real-time distances to nearby individuals, illustrating this feature. In addition, randomness runs through all aspects of the model, so that different runs with identical parameters often produce different outcomes. Finally, there is adaptability and feedback: agents continuously adjust their actions in response to the environment. In the Schelling model, the movement of one agent may cause its neighbors to become dissatisfied, triggering a new round of relocation, and the system’s complexity evolves through this ongoing cycle (Schelling, 1971).

Before Schelling, it was widely assumed that strong spatial segregation must reflect strong individual prejudice. Schelling's model challenged this assumption, shifting researchers' attention from individual attitudes to the structural dynamics of collective behaviour(Schelling, 1971; Bruch and Mare, 2006).

The model omits many of the factors that shape actual segregation: income inequality, housing markets, historical discrimination, and institutional racism(Massey, 1990). None of these fall within the model's scope. The model's strength lies in revealing a mechanism, not in simulating a reality(O’Sullivan, 2004).


    
2. What are the key ABM properties that the Schelling ABM showcases? 
    - Consider concepts such as emergence, complexity, interactions, behaviours... etc.
    
    The Schelling model is almost a textbook example of the core properties of ABMs. The most prominent of these is emergence: no agent is instructed to produce segregation, each agent simply evaluates its neighbourhood and moves if dissatisfied, yet the result is a highly clustered, sharply divided spatial distribution(Schelling, 1971). More intriguingly, the degree of segregation that ultimately emerges often far exceeds the similarity threshold that agents themselves require — the outcome of the system can surprise even its designers(Bruch and Mare, 2006).
    
    Underpinning emergence is the combined effect of local interactions and feedback. Each agent's decision to stay or move depends entirely on its eight immediately neighbouring cells, with no knowledge of the overall state of the grid(Schelling, 1971). At the same time, an agent's movement alters the neighbourhood composition of surrounding cells, which may trigger previously satisfied neighbours to become dissatisfied and relocate in turn. Such cascading effects produce non-linear dynamics that would be difficult to anticipate without the aid of simulation.
    
    In addition, randomness and heterogeneity are equally important. Two runs with identical parameters will produce different spatial patterns, yet macroscale statistics tend to converge, revealing the role of path dependency. The fact that agents are of only two types demonstrates that even the most basic individual differences are sufficient to give rise to complex collective outcomes(Bonabeau, 2002).

    The policy implications of the Schelling model are clear: even if a policy temporarily makes a community more diverse, residents near the tipping point may immediately begin to move, triggering a cascade effect that pushes the community back toward segregation(Schelling, 1971). This suggests that stable integration requires sustained intervention rather than a one-time mixing of populations. The model also reveals that similarity preferences have a tipping point, below which the system remains integrated and above which segregation rapidly emerges, meaning that small changes in social attitudes may produce disproportionately large long-term consequences. More profoundly, segregation is a product of system dynamics rather than a direct reflection of individual prejudice — individuals need not harbour malicious intent for extreme spatial separation to emerge(Bruch and Mare, 2006). As overt discrimination gradually declines, what structural conditions continue to sustain segregation remains a question worthy of deeper investigation.

Part 3: Critically Evaluate the Schelling model (is it fit for purpose?)

The greatest strength of the Schelling model is its economy. Two agent types, one behavioural rule, and one key parameter are sufficient to produce a result of genuine intellectual significance: mild individual preferences can give rise to strong collective segregation(Schelling, 1971). Its contribution also lies in its capacity to surprise — before Schelling, it was almost universally assumed that strong segregation implied strong individual prejudice, and the model demonstrates that this inference is not logically necessary. Furthermore, decades after its publication, the model continues to be extended to domains such as school segregation, occupational sorting, and opinion polarisation, consistently generating productive research questions(Bruch and Mare, 2006).

The limitations are equally significant. The model compresses human decision-making into a single binary rule: stay if satisfied, relocate randomly if not. In reality, relocation decisions are driven by numerous factors including income, commuting distance, social networks, and cultural ties. More critically, the model treats relocation as costless and random, entirely disregarding the structural constraints imposed by economic inequality. This risks producing a misleading narrative that obscures the deeper role of discriminatory housing policies and wealth disparities. The model also represents society as two symmetric groups, failing to reflect the asymmetries of power in real cities. It has no memory, and therefore cannot account for how the legacy of historical discrimination persists to the present day. Agents have no knowledge of macroscale institutional structures, whereas in reality government policy and property values actively shape individual choices(O’Sullivan, 2004).

As a theoretical demonstration of a social mechanism, the Schelling model is unquestionably fit for purpose: it proves that mild homophily is a sufficient condition for macroscale segregation to emerge, which is a genuine and important insight. However, as a basis for policy design, the model's omission of economic, historical, and institutional factors renders it insufficient(O’Sullivan, 2004). The model demonstrates that segregation can arise from mild preferences alone, but does not demonstrate that real-world segregation does in fact arise primarily from this mechanism. Conflating the two is a significant interpretive error(Bruch and Mare, 2006).
